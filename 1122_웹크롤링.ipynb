```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

import urllib.request
import urllib.parse

from bs4 import BeautifulSoup
```

```python
try:
    html=urlopen("http://jaaaaaa.com")
except HTTPError as e:
    print("HTTP 에러입니다.")
except URLError as e:
    print("존재하지 않는 사이트입니다.")
else:
    print(html.read())
```

```python
url="http://cdn.newsen.com/newsen/news_photo/2019/11/22/201911220200091810_1.jpg"
savename="data/images/test02.jpg"
image=urllib.request.urlopen(url).read()
with open(savename,mode="wb")as f:
    f.write(image)
    print("저장되었습니다.")
```

```python
#매개변수를 추가하여 인터넷 리소스를 요청하는 방법
API="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
values={'stnld':'159'}
params=urllib.parse.urlencode(values)

url=API+"?"+params

print("url=",url)

data=urllib.request.urlopen(url).read()
text=data.decode('utf-8')
print(text)
```

```python
html=urlopen("http://stackoverflow.com")
bs=BeautifulSoup(html.read(),'html.parser')
print(bs.title)
print(bs.title.text)
print(bs.h1)
print(bs.h1.text)
print(bs.span)
```

```python
html='''
<html>
<body>
<h1>hello python</h1>
<p>웹페이지 복사</p>
<p>웹 스크래핑</p>
<h1>hello java</h1>
</body>
</html>

'''
bs=BeautifulSoup(html,'html.parser')
h1=bs.html.body.h1
p1=bs.html.body.p

print(h1.string)
print(p1.text)

p2=p1.next_sibling.next_sibling
print(p2)
h2=h1.next_sibling.next_sibling
print(h2)


h1=bs.find('h1')
print(h1.text)

```

```python
html = urlopen("http://news.naver.com/")
bs=BeautifulSoup(html.read(),'html.parser')
links=bs.find_all('a')

for link in links:
    print(link.attrs['href'])
    print(link.string)
```
